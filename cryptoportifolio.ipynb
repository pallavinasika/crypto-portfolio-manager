Generate a Python program that demonstrates the use of functions to calculate profit for a crypto asset. The function should accept buy_price, current_price, and quantity as parameters and return the profit.

```python
def calculate_profit(buy_price, current_price, quantity):
    profit = (current_price - buy_price) * quantity
    return profit

# Example usage
buy_price = 30000
current_price = 68000
quantity = 0.5

profit = calculate_profit(buy_price, current_price, quantity)

print("Profit:", profit)
```

Generate a Python program that stores crypto assets in a list of dictionaries and calculates total investment using a loop.

```python
portfolio = [
    {"coin": "bitcoin", "quantity": 0.5, "buy_price": 30000},
    {"coin": "ethereum", "quantity": 2, "buy_price": 2000}
]

total_investment = 0

for asset in portfolio:
    investment = asset["quantity"] * asset["buy_price"]
    total_investment += investment
    print(f"{asset['coin']} Investment:", investment)

print("Total Investment:", total_investment)
```

Generate a Python class named CryptoAsset with attributes name, quantity, and buy_price. Include a method to calculate investment.

```python
class CryptoAsset:
    def __init__(self, name, quantity, buy_price):
        self.name = name
        self.quantity = quantity
        self.buy_price = buy_price

    def calculate_investment(self):
        return self.quantity * self.buy_price


btc = CryptoAsset("bitcoin", 0.5, 30000)
eth = CryptoAsset("ethereum", 2, 2000)

print("BTC Investment:", btc.calculate_investment())
print("ETH Investment:", eth.calculate_investment())
```

Generate a Python program that finds the crypto asset with the highest investment from a list of assets.

```python
portfolio = [
    {"coin": "bitcoin", "quantity": 0.5, "buy_price": 30000},
    {"coin": "ethereum", "quantity": 2, "buy_price": 2000},
    {"coin": "solana", "quantity": 10, "buy_price": 100}
]

max_investment = 0
max_coin = ""

for asset in portfolio:
    investment = asset["quantity"] * asset["buy_price"]

    if investment > max_investment:
        max_investment = investment
        max_coin = asset["coin"]

print("Highest Investment Coin:", max_coin)
print("Investment Value:", max_investment)
```

Generate an SQL query to create a table named portfolio with id (primary key), coin_name (text), quantity (real), and buy_price (real).

```python
import sqlite3

conn = sqlite3.connect("portfolio.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS portfolio (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    coin_name TEXT,
    quantity REAL,
    buy_price REAL
)
""")

conn.commit()
print("Table created successfully")
```

Generate SQL queries to insert two crypto assets into the portfolio table.

```python
cursor.execute("""
INSERT INTO portfolio (coin_name, quantity, buy_price)
VALUES ('bitcoin', 0.5, 30000)
""")

cursor.execute("""
INSERT INTO portfolio (coin_name, quantity, buy_price)
VALUES ('ethereum', 2, 2000)
""")

conn.commit()
print("Data inserted successfully")
```

Generate an SQL query to retrieve all records from the portfolio table.

```python
cursor.execute("SELECT * FROM portfolio")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

Generate an SQL query to calculate total investment using SUM(quantity * buy_price).

```python
cursor.execute("""
SELECT SUM(quantity * buy_price) AS total_investment
FROM portfolio
""")

total = cursor.fetchone()[0]
print("Total Investment:", total)
```

Generate an SQL query to update the quantity of bitcoin in the portfolio table.

```python
cursor.execute("""
UPDATE portfolio
SET quantity = 1
WHERE coin_name = 'bitcoin'
""")

conn.commit()
print("Record updated")
```

Generate an SQL query to delete ethereum from the portfolio table.

```python
cursor.execute("""
DELETE FROM portfolio
WHERE coin_name = 'ethereum'
""")

conn.commit()
print("Record deleted")
```
